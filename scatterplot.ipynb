```python
import matplotlib.pyplot as plt
import numpy as np

# Data
years = np.array([2018, 2019, 2020, 2021, 2022, 2023])
rainfall = np.array([820, 760, 890, 920, 870, 940])
yield_production = np.array([210, 195, 240, 260, 250, 275])

# Line Plot - Rainfall Trend
plt.figure(figsize=(8,5))
plt.plot(years, rainfall, marker='o')
plt.title("Yearly Rainfall Trend")
plt.xlabel("Year")
plt.ylabel("Rainfall (mm)")
plt.grid(True)
plt.show()

# Scatter Plot - Rainfall vs Yield
plt.figure(figsize=(8,5))
plt.scatter(rainfall, yield_production)
plt.title("Rainfall vs Crop Yield")
plt.xlabel("Rainfall (mm)")
plt.ylabel("Yield (tons)")
plt.grid(True)
plt.show()
```
