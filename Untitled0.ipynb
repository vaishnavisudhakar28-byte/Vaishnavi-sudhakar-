```python
season = input("Enter season (Kharif/Rabi): ").lower()
rainfall = int(input("Enter rainfall in mm: "))
if season == "kharif":
	if rainfall > 600:
		print("Grow Rice")
	else:
		print("Grow Pulses")
else:
	print("Grow Wheat")
```
