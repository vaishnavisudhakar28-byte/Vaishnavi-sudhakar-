```python
car = {
"brand": "Ford",
"model": "Mustang",
"year": 1964
}

x = car.values()
y=len(car)
print(y)

print(x) #before the change

car["year"] = 2020
y=len(car)
print(y)

print(x) #after the change
```

```python
car = {
"brand": "Ford",
"model": "Mustang",
"year": 1964
}
x = car.values()
print(x) #before the change
car["color"] = "red"
print(x) #after the change
```

```python
car = {
"brand": "Ford",
"model": "Mustang",
"year": 1964
}

x = car.items()

print(x) #before the change

car["color"] = "red"

print(x) #after the change
```
