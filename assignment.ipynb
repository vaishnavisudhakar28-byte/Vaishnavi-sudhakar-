```python
#lab assignment 8 :Average rainfall analysis
def calculate_average_rainfall(rainfall_data):
    total_rainfall = sum(rainfall_data)
    average_rainfall = total_rainfall / len(rainfall_data)
    return average_rainfall
```

```python
#labassignment 8: average rainfall analysis
rainfall=[]
for i in range(5):
  value=float(input(f"enter rainfall for day{i+1}:"))
  rainfall.append(value)
# Calculate average after collecting all inputs
average = sum(rainfall) / 5
print("average rainfall:",average)
if average>=100:
  print("sufficient rainfall")
else:
  print("insufficient rainfall")
```
