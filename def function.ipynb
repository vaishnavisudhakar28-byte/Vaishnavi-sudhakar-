```python
def is_ready_for_harvest(days_since_planted,plant_type):
    if days_since_planted >= 120 and plant_type=="Rice":
        return True
    else:
        return False
plant_type=input("Enter the plant type")
days_since_planted=int(input("Enter the number of days since planted"))
is_ready_for_harvest(days_since_planted,plant_type)
```

```python
#Returning Multiple Values

def weather_conditions():
    temperature = 28  # in Celsius
    humidity = 70  # in percentage
    return temperature, humidity  # returns a tuple

# Usage
temp, hum =weather_conditions()
print(f"Temperature: {temp}C, Humidity: {hum}%")
```
