```python
rainfall = [450, 800, 1200]
for r in rainfall:
    if r < 500:
        print(r, "→ Low rainfall")
    elif r < 1000:
        print(r, "→ Moderate rainfall")
    else:
        print(r, "→ Sufficient rainfall")
```
