```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Define tenors in days
tenors = [
    7, 15, 30, 45, 60, 90, 120, 180, 270, 365, 540, 730
]

tenor_labels = [
    '7 Days', '15 Days', '1 Month', '45 Days', '2 Months', '3 Months',
    '4 Months', '6 Months', '9 Months', '1 Year', '1.5 Years', '2 Years'
]

# Generate synthetic interest rates
# Assuming a base rate and a slight increase for longer tenors, plus some noise
np.random.seed(42)
base_rate = 3.0
interest_rates = []

for i, tenor in enumerate(tenors):
    # Simulate a general trend where longer tenors have slightly higher rates
    # Add some random fluctuation
    rate = base_rate + (i * 0.2) + np.random.uniform(-0.5, 0.5)
    # Ensure rates are positive and within a reasonable range (e.g., 2% to 7%)
    rate = max(2.0, min(rate, 7.0))
    interest_rates.append(rate)

fd_data = pd.DataFrame({
    'Tenor_Days': tenors,
    'Tenor_Label': tenor_labels,
    'Interest_Rate': interest_rates
})

print("Generated FD Interest Rate Data:")
display(fd_data)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.lineplot(x='Tenor_Label', y='Interest_Rate', data=fd_data, marker='o')
plt.title('FD Interest Rates by Tenor')
plt.xlabel('Tenor')
plt.ylabel('Interest Rate (%)')
plt.xticks(rotation=45, ha='right')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.barplot(x='Tenor_Label', y='Interest_Rate', data=fd_data, palette='viridis')
plt.title('FD Interest Rates by Tenor (Bar Graph)')
plt.xlabel('Tenor')
plt.ylabel('Interest Rate (%)')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
