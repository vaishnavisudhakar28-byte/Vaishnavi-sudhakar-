```python
def irrigation_needed (moisture_level,rainfall):
  if moisture_level <=40 and rainfall <= 10:
    print("Irrigation needed")
  else:
    print("No irrigation needed")
```

```python
irrigation_needed(40,10)
```

```python
irrigation_needed(50,30)
```
